## Enter 9 to see the analysis of Number of Orders Across Parts of a Day

```python
orders_df = pd.read_csv('./orders 2020-2021.csv')
time = orders_df[['Fulfillment Date and Time Stamp']]
time = time.dropna()
times = []
```

```python
for i in range(time.shape[0]):
        times.append(time['Fulfillment Date and Time Stamp'].iloc[i].split(" ")[1].split(":")[0])
for i in range(len(times)):
        times[i] = int(times[i])
morning = []
```

```python
for i in range(len(times)):
        if times[i] > 5 and times[i] < 12:
            morning.append(times[i])
afternoon = []
for i in range(len(times)):
        if times[i] >= 12 and times[i] < 16:
            afternoon.append(times[i])
evening = []
```

```python
for i in range(len(times)):
        if times[i] >= 16 and times[i] < 22:
            evening.append(times[i])
night = []
for i in range(len(times)):
        if times[i] >= 22 and times[i] <= 23:
            night.append(times[i])
        if times[i] >= 0 and times[i] <= 5:
            night.append(times[i])
```

```python
def fun9():
 import xlrd, xlwt
 wb = Workbook()
 wb = xlwt.Workbook()
 sheet1 = wb.add_sheet('Sheet 1')
 style = xlwt.easyxf('font: bold 1')
 sheet1.write(0, 0, 'Session', style)
 sheet1.write(0, 1, '#Orders', style)
 sheet1.write(1, 0, "Morning")
 sheet1.write(2, 0, "Afternoon")
 sheet1.write(3, 0, "Evening")
 sheet1.write(4, 0, "Night")
 sheet1.write(1, 1, len(morning))
 sheet1.write(2, 1, len(afternoon))
 sheet1.write(3, 1, len(evening))
 sheet1.write(4, 1, len(night))
 wb.save('NumberOfOrderAcrossDay.xls')
```
